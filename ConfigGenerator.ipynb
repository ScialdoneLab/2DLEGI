# Generation of 2D configurations of epithelial cells

```python
%matplotlib inline
```

Here we define a generator of a planar epithelial sheet. Given the number of cells, sheets with different mean polygon number (i.e., average number of sides of each cell) are generated, starting from a standard epithelial sheet (hexagonal cells) and randomly collapsing edges in an iterative way.

```python
import numpy as np
import pandas as pd
import ipyvolume as ipv
import matplotlib
import matplotlib.pylab as plt
import matplotlib.pyplot as plt

from tyssue import config
from tyssue import Sheet
from tyssue import PlanarGeometry as geom
from tyssue.dynamics import model_factory, effectors
from tyssue.solvers.quasistatic import QSSolver

from tyssue.topology.base_topology import collapse_edge, remove_face

from tyssue.draw import sheet_view, highlight_cells

from tyssue.config import draw

from tyssue.utils import connectivity
from tyssue.io import hdf5

import random
import warnings
```

```python
warnings.filterwarnings('ignore')
from sklearn.cluster import KMeans

def BasicSheet():
    s = Sheet.planar_sheet_2d('flat', 40, 40, 10, 10, noise=0.5)
    
    # Cut the sheet to reduce the number of cells
    to_cut = s.cut_out([(1, 140), (1, 140)])
    s.remove(to_cut, trim_borders=False)
    
    # Update sheet properties
    geom.update_all(s)
    s.update_rank()
    
    return s;

def FindCentroid(s):
    
    # Cell coordinates
    x_coords=np.array(s.face_df['x'])
    y_coords=np.array(s.face_df['y'])
    xy_coords=np.column_stack((x_coords,y_coords))
    
    # Compute centroid
    kmeans_s = KMeans(n_clusters=1, random_state=0).fit(xy_coords)
    centre_s = kmeans_s.cluster_centers_[0]
    
    return centre_s,xy_coords;

def ClosestCell(s):
    
    centre_s,xy_coords=FindCentroid(s)
    
    # Distance of each cell from the centroid
    dist_centroid=np.sqrt( (xy_coords[:,0] - centre_s[0])**2 + (xy_coords[:,1] - centre_s[1])**2 )
    
    # Find cell closest to the configuration centroid
    idx_min=np.argmin(dist_centroid)
    
    return idx_min;

def SelectCells(sheet,c_lst):
    my_cells=[]
    
    for i in c_lst:
        my_cells.append(list(sheet.get_neighbors(i)))
    flat_list = list(c_lst) + [item for sublist in my_cells for item in sublist]
    return set(flat_list); 

def SubsetConf(s,idx):
    new_s=s.copy()
    edges_to_keep = set(new_s.edge_df[new_s.edge_df['face'].isin(idx)].index)
    edges_to_cut=pd.Int64Index(set(new_s.edge_df.index)-set(edges_to_keep),name='edge')
    new_s.remove(edges_to_cut, trim_borders=False)
    geom.update_all(new_s)
    new_s.update_rank()
    
    return new_s;

def PlotSheet(s):
    print(s.identifier)
    print('Nr cells:', s.Nf)
    print('Mean Polygon Number',s.face_df['num_sides'].mean())
    fig, ax = sheet_view(s, mode="quick")
    plt.show(),plt.close()
    

# Generate NC epithelial configurations with hexagonal cells
def ConfGen(NC):
    
    # Define a large sheet of hexagonal cells
    s=BasicSheet()
    
    # Find the cell closest to the configuration centroid
    idx_min=ClosestCell(s)
    new_idx_min=[idx_min]
    cell_lists=[]
    my_sheets=[]
    
    for i in range(NC):
        new_idx_min=SelectCells(s,new_idx_min)
        new_s=SubsetConf(s,new_idx_min)
        new_s.identifier='hexa'+str(i)
        cell_lists.append(list(new_idx_min))
        
        min_x = new_s.vert_df['x'].min()
        min_y = new_s.vert_df['y'].min()
        
        new_s.vert_df['x'] -= abs(min_x)
        new_s.vert_df['y'] -= abs(min_y)
        geom.update_all(new_s)
        my_sheets.append(new_s)
        PlotSheet(new_s)
        
    return my_sheets;

def DefineModel(s):
    # Define a mechanical model associated to the sheet
    model = model_factory(
        [
        effectors.LineTension,
        effectors.FaceContractility,
        effectors.FaceAreaElasticity
        ]
    )
    
    specs = {
        "face": {
        "contractility": 5e-2,
        "prefered_area": s.face_df.area.mean(),
        "area_elasticity": 1.
        },
        "edge": {
        "line_tension": 1e-2,
        "is_active": 1
        },
        "vert": {
        "is_active": 1
        },
    }
    
    solver = QSSolver()
    
    return model,specs,solver;

def MidPoints(s):
    s.edge_df['midx']=0.5*(s.edge_df['sx']+s.edge_df['tx'])
    s.edge_df['midy']=0.5*(s.edge_df['sy']+s.edge_df['ty'])

# Find the edges available for collapse (those connected to vertices with rank <= 4)
def FindAvEdges(s):
    # Find vertices with rank >=4
    high_rank_vertices=list(s.vert_df[s.vert_df['rank']>3].index)
    
    # Find edges connected to the vertices with rank >=4
    edges_high_rank_vertices=list(s.edge_df[s.edge_df['srce'].isin(high_rank_vertices) | s.edge_df['trgt'].isin(high_rank_vertices)].index)
    
    # Remove those edges from the dataframe
    subset_edge_df=s.edge_df.drop(edges_high_rank_vertices)
    
    # Return indices and lengths of the remaining edges
    av_edges=list(subset_edge_df.index)
    av_edges_l=list(subset_edge_df['length'])
    
    return av_edges, av_edges_l;

def find_nearest(array,MPNf,NS):
    values=np.linspace(array[0],MPNf,num=NS,endpoint=True)
    
    indices=[]
    
    for i in values:
        idx = (np.abs(array - i)).argmin()
        indices.append(idx)
        
    return indices;
    
def CollapseEdges(s_old,T_MPN,T_NR=0):
    
    tmp_sheets=[]
    tmp_MPN=[]
    
    s=s_old.copy()
    
    tmp_sheets.append(s_old)
    tmp_MPN.append(s.face_df['num_sides'].mean())
    model,specs,solver=DefineModel(s)
    
    s.update_specs(specs, reset=True)
    
    # Find the ground state
    res = solver.find_energy_min(s, geom, model)
    
    # Compute configuration centroid
    centre_s,_=FindCentroid(s)
    
    # Find midpoints of the edges
    MidPoints(s)
    
    # Distance of each edge midpoint from the configuration centroid
    s.edge_df['dist']=np.sqrt( (s.edge_df['midx'] - centre_s[0])**2 + (s.edge_df['midy'] - centre_s[1])**2 )
    
    if T_NR>0:
        print('\033[1m' + '2. Starting rosettes\' generation')
        # First generate the target number of rosettes
        s=RosetteFormation(s,T_NR,model,specs,solver)
        print('\033[1m' + '2. %d multi-cellular rosettes have been generated' % (T_NR))
    else:
        print('\033[1m' + '2. Target rosette nr = 0: Skipping rosettes\' generation')
    
    # Compute the mean polygon number
    MPN=s.face_df['num_sides'].mean()
    
    # Collapse edges until the target mean polygon number is reached
    # (without selecting the edges belonging to the rosettes and keeping the vertex rank <=4)
    print('\033[1m' +'Current mean polygon number: %f.' % (MPN))
    print('\033[1m' +'Target mean polygon number: %f.\n' % (T_MPN))
    
    print('\033[1m' +'3. Starting iterative edge collapse to reach the target mean polygon number')
    while MPN > T_MPN:
        # Choose an edge
        edge_to_collapse = random.choices(s.edge_df.index, k=1)[0]
        collapse_edge(s, edge_to_collapse)
        s.update_rank()
        geom.update_all(s)
        res = solver.find_energy_min(s, geom, model)
        MPN=s.face_df['num_sides'].mean()
        
        # Compute configuration centroid
        centre_s,_=FindCentroid(s)
        
        # Find midpoints of the edges
        MidPoints(s)
        
        # Distance of each edge midpoint from the configuration centroid
        s.edge_df['dist']=np.sqrt( (s.edge_df['midx'] - centre_s[0])**2 + (s.edge_df['midy'] - centre_s[1])**2 )
    
        min_x = s.vert_df['x'].min()
        min_y = s.vert_df['y'].min()
        
        s.vert_df['x'] -= np.sign(min_x)*abs(min_x)
        s.vert_df['y'] -= np.sign(min_y)*abs(min_y)
        geom.update_all(s)
        tmp_s=s.copy()
        tmp_sheets.append(tmp_s)
        tmp_MPN.append(s.face_df['num_sides'].mean())
  
    return tmp_sheets,tmp_MPN;
```
